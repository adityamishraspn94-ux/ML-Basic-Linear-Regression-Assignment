## **Q1. What is Machine Learning? Explain the main types of Machine Learning with examples.**

## **Answer:**

Machine Learning (ML) is a branch of Artificial Intelligence (AI) that enables computers to learn patterns from data and make predictions or decisions without being explicitly programmed.

Machine Learning systems improve their performance automatically through experience and data.

## **Main Types of Machine Learning**

### **1. Supervised Learning**

In supervised learning, the model is trained using labeled data, where both input and output are known.

### **Examples:**

* House price prediction
* Email spam detection
* Student result prediction

### **Algorithms:**

* Linear Regression
* Logistic Regression
* Decision Tree

### 2. **Unsupervised Learning**

In unsupervised learning, the model works with unlabeled data and tries to find hidden patterns or groups.

### **Examples:**

* Customer segmentation
* Market basket analysis

### **Algorithms:**

* K-Means Clustering
* Hierarchical Clustering

### **3. Reinforcement Learning**

In reinforcement learning, an agent learns by interacting with the environment and receiving rewards or penalties.

### **Examples:**

* Self-driving cars
* Game-playing AI
* Robotics

# **Q3. What is Linear Regression? Write the Linear Regression equation and explain intercept and coefficient.**

## **Answer:**

Linear Regression is a supervised machine learning algorithm used to predict continuous numerical values based on the relationship between dependent and independent variables.

It tries to fit the best straight line through the data points.

## **Linear Regression Equation**

y=b_0+b_1x_1+b_2x_2+\cdots+b_nx_n

Where:

* (y) = dependent variable
* (b_0) = intercept
* (b_1, b_2) = coefficients
* (x_1, x_2) = independent variables

## **Intercept ((b_0))**

The intercept is the predicted value of (y) when all independent variables are zero.

## **Coefficient ((b_1, b_2))**

Coefficients represent how much the dependent variable changes when one independent variable increases by one unit while keeping others constant.

**Example:**
If coefficient of RM = 3, then increasing one room increases house price by 3 units.

## **Q4. Explain the following evaluation metrics used in Linear Regression.**

## **1. SSE (Sum of Squared Errors)**

SSE measures the total squared difference between actual and predicted values.

SSE=\sum (y_i-\hat{y}_i)^2

Lower SSE indicates better model performance.

## **2. MSE (Mean Squared Error)**

MSE is the average of squared errors.

MSE=\frac{1}{n}\sum (y_i-\hat{y}_i)^2

Lower MSE means better accuracy.

## **3. RMSE (Root Mean Squared Error)**

RMSE is the square root of MSE.

RMSE=\sqrt{\frac{1}{n}\sum (y_i-\hat{y}_i)^2}

It measures prediction error in the same unit as the target variable.

## **4. R² Score**

R² Score measures how well the model explains the variance in the target variable.

[
R^2 = 1 - \frac{SSE}{SST}
]

Range:

* 0 to 1
* Higher value indicates better model performance

## **5. Adjusted R²**

Adjusted R² improves R² by considering the number of predictors used in the model.

It prevents unnecessary variables from increasing model accuracy artificially.

**Formula:**

[
Adjusted\ R^2 = 1 - \left( \frac{(1-R^2)(n-1)}{n-p-1} \right)
]

Where:

* (n) = number of observations
* (p) = number of predictors

## **Q5. Explain the Machine Learning workflow steps from data collection to model evaluation.**

## **Answer:**

The Machine Learning workflow consists of several important steps:

## **1. Data Collection**

Data is collected from databases, websites, sensors, surveys, or files.

**Example:**
Housing price dataset.

## **2. Data Cleaning**

Incorrect, duplicate, or missing values are handled.

Tasks include:

* Removing null values
* Handling outliers
* Formatting data

## **3. Exploratory Data Analysis (EDA)**

EDA helps understand patterns and relationships in data using:

* Charts
* Heatmaps
* Pairplots
* Summary statistics

## **4. Feature Selection**

Important variables are selected for model building.

**Example:**
Rooms, crime rate, tax rate.

## **5. Data Splitting**

Dataset is divided into:

* Training data
* Testing data

Common split:
80% training and 20% testing.

## **6. Model Building**

A machine learning algorithm is trained using training data.

**Example:**
Linear Regression model.

## **7. Model Prediction**

The trained model predicts values using test data.

## **8. Model Evaluation**

Performance is checked using metrics such as:

* MAE
* MSE
* RMSE
* R² Score

## **9. Model Improvement**

**Techniques like:**

* Feature engineering
* Hyperparameter tuning
* Removing irrelevant variables

are used to improve model performance.
